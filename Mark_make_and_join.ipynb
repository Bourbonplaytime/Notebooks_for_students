```python
DIDs = list(df['DIDs '])
```

```python
new_df = pd.DataFrame(DIDs)
```

```python
new_df
```

```python
new_df['random'] = random.rand(100)
```

```python
def grouping(i):
    if i > 0.5:
        result = 'S'
    elif i <= 0.5:
        result = 'L'
    return result
```

```python
new_df['Grouping'] = new_df['random'].map(lambda x: grouping(x))
                
```

```python
new_df.columns = ['DIDs ', 'Randnum', 'Grouping']

new_df
```

```python
new_df.to_csv('c:/users/bourb/downloads/dummy_data.csv', index=False)
```
